```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python

# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://dashboard.animaltrust.org.uk/uploads/What_To_Do_When_Your_Cat_Is_In_Heat_72e4a61bc6.png"]
#baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list for url in urls:
for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
#read image
#resize image to half size #convert color to RGB #concatenate image #show image
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi, " x ",lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image for y in range (lebar):
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range (tinggi):
  image_5[x, int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_4, image_5))
cv2_imshow(final_frame)
```

```python
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
m = float(tinggi - 1) / (lebar - 1)

for i in range (lebar):
    row1 = int(m * i)
    image_6[row1, i] = [255, 255, 255]
    row2 = (tinggi - 1) - int(m * i)
    image_6[row2, i] = [255, 255, 255]

final_frame = cv.hconcat((image_4, image_6))
cv2_imshow(final_frame)
```

```python
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

size = int(tinggi * 0.40)
cord_y = (tinggi // 2) - (size // 2)
cord_x = (lebar // 4) - (size // 2)

for y in range (cord_y, cord_y + size):
  for x in range (cord_x, cord_x + size):
      image_7[y,x] = [255,255,255]


final_frame = cv.hconcat((image_4, image_7))
cv2_imshow(final_frame)
```
